### Interacting with LLM

```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    base_url="http://localhost:11434",
    model="qwen2.5:latest",
    temperature=0.2,
    max_tokens=250
)

#invoking an LLM -> use the llm variable to call invoke method
response = llm.invoke("Hi, What is your name?")
print(response)
```

#### PROMPT and CHAT Template

```python
from langchain_core.prompts import PromptTemplate

## setup a template for input prompt --> use PromptTemplate
prompt_template = PromptTemplate.from_template("What is the capital of {country_name} and what is the famous tourist spot?")
country_input=input()
prompt = prompt_template.invoke({"country_name":country_input})

print(prompt_template)
#content = llm.invoke(prompt).content
#print(content)
```

#### CHAT PROMPT TEMPLATE - to set the llm as a role

```python
from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate, SystemMessagePromptTemplate

system_msg = SystemMessagePromptTemplate.from_template("You are an LLM expert")
human_msg = HumanMessagePromptTemplate.from_template("How will you setup Guardrails to a {project_type} project?")
prompt_chat_template = ChatPromptTemplate([system_msg,human_msg])

# 2. Option 2 of doing the same
#prompt_chat_template = ChatPromptTemplate([
 #   ("system","You are an LLM expert"),
 #   ("user","How will you setup Guardrails to a {project_type} project?")
#])

prompt = prompt_chat_template.invoke({"project_type":"test automation"})
content = llm.invoke(prompt).content
#print(prompt_chat_template)
print(content)
```

#### MESSAGE PLACEHOLDER --> 

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import HumanMessage

prompt_template3 = ChatPromptTemplate([
    ("system","You are a tourist guide in Italy"),
    MessagesPlaceholder("msg")
])

prompt3 = prompt_template3.invoke({"msg":[HumanMessage("What is the first place you will take a tourist to? Just give the name")]})

#content3 = llm.invoke(prompt3).content
# invoke vs stream
for st in llm.stream(prompt3):
    print(st.content3)

```
